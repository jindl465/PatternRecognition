# Import packages numpy and matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Define a differentiable function that maps from real number to real number.

## Define $f(x) = x * sinx + x * cosx $ as func(x)

```python
def func(x):
    f =  x * np.sin(x) + x * np.cos(x)
    return f
```

# Define a domain of the function.

## Domain : $-8<x<8$

```python
x = np.arange(-8,8,0.1)
```

# Plot the function.

```python
f = func(x)
plt.figure(1)
plt.plot(x, f, 'k', label="function")
```

# Select a point within the domain.

## Point $x$ : -3, 4 

```python
p1 = -3
p2 = 4
```

## Mark the selected point on the function.

```python
def origin_plot():
    y1 = func(p1)
    y2 = func(p2)

    plt.scatter(p1,y1,c ='r',s = 30)
    plt.scatter(p2,y2, c ='b',s = 30)
    plt.plot(x,f,'k',label="function")

origin_plot()
```

# Define the first-order Taylor approximation at the selected point.

## Define $f'(x) = sinx + x*cosx + cosx - x*sinx$ as derivate function d_func(x)

```python
def d_func(x):
    df = np.sin(x) + x*np.cos(x) + np.cos(x) - x*np.sin(x)
    return df
```

## Define Tylor Approximation $f(p) + f'(p)(x-p)$

```python
def tylor(p,x):
    result = func(p) + d_func(p) * (x-p)
    return result
```

# Plot the Tylor approximation with the same domain of the original function.

```python
origin_plot()

df1 = tylor(p1,x)
df2 = tylor(p2,x)
plt.plot(x, df1, 'r')
plt.plot(x, df2, 'b')
```
